```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup,Tag
from selenium import webdriver
import requests
from datetime import datetime,date
```

```python
driver=webdriver.Chrome(executable_path='C:/Users/admin/chromedriver.exe')
url='https://www.mohfw.gov.in/'
driver.get(url)
webContent=driver.page_source
pysoup=BeautifulSoup(webContent,'html')
driver.quit()
```

```python
divdata=pysoup.find('div',{'class':'data-table table-responsive'})
divdate=divdata.h5.span.text.split(',')[0].split(':')[1].strip()
divdate
formate='%d %B %Y'
datetimeobj=datetime.strptime(divdate,formate).date()
datetimeobj
```

```python
tr_head=pysoup.find('thead').find('tr',{'class':'row2'})
tr_head
headings=[th.text.strip() for th in tr_head.find_all('th')]
headings=['S. No.','Name of State / UT','Total',
 'Change since yesterdayChange since yesterday',
 'Cumulative',
 'Change since yesterday',
 'Cumulative',
 'Change since yesterday']
```

```python
listoftr=pysoup.find('tbody').find_all('tr')
mytable=[]
for tr in listoftr:
    if isinstance(tr,Tag):
        stateData=[-int(td.text.strip())\
                   if td.span != None and td.span.get('class') in [['down']]\
                   else td.text.strip()
                   for td in tr.find_all('td')]
        data=dict(zip(headings,stateData))
        mytable.append(data)
       
state_data=pd.DataFrame(mytable,columns=headings)    
state_data
```
